```python
mj.plot(kind='scatter',x='PPG',y='Year',color='red', label = 'Points per game')


plt.title ('Michael Jordan Scoring Average')
plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
plt.show() 
#As we can see, Michael Jordan lowest scoring avarage was in 2002 and his highest scoring avarage was in 1987. 
#Even at 40 years old, Michael Jordan was averaging 20 points which is pretty impressive.
```

```python
'''How long did it take for MJ and Kobe to average their highest scoring average?'''
ax = plt.gca() # gca stands for 'get current axis'

mj.plot(kind='line',x='PPG',y='Year', color='red', label ='Michael Jordan', ax=ax)
kb.plot(kind='line',x='PPG',y='Year', color='purple', label = "Kobe Bryant", ax=ax)


plt.xlabel('Points Per Game')
plt.ylabel('Year')

plt.title ('Highest scoring Average, How long it took for them to hit that peak?')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
plt.show()

'''By Kobe 10th season he averaged his highest scoring averaged, 35 in 2006. While
Michael Jordan was averaging 37 points in just his 3rd Season! Lets not
forget he was comming back from a broken foot. (Jordan had a broken foot in 1986.'''
```

```python
ax = plt.gca() # gca stands for 'get current axis'

bird.plot(kind='scatter',x='AST',y='Year', color='green', label ='Larry Bird', ax=ax)
lb.plot(kind='scatter',x='AST',y='Year', color='blue', label = "Lebron James", ax=ax)
vc.plot(kind='scatter',x='AST',y='Year', color='purple', label = "Vince Carter", ax=ax)


plt.xlabel('Assist Per Game')
plt.ylabel('Season Year')

plt.title ('Who averaged more assist per game in their careers?')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
plt.show()
#As You can see, Lebron is the greastest Assist maker in the small foward position, 
#he averages about 9 assit a game in the 2019 season compare to the bird man who only averaged
#7 assist a game at most and vince carter with about 5 assist a game in 2009 at most.''' 
```

```python
ax = plt.gca() # gca stands for 'get current axis'
mj.plot(kind='bar',x='Year',y='PPG',color='red', label ='Michael Jordan', ax=ax)

plt.xlabel('Season Year')
plt.ylabel('Points Per Game')

plt.title ('Air Jordan Career Points Stats')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
```

```python
ax = plt.gca() # gca stands for 'get current axis'
lb.plot(kind='bar',x='Year',y='PPG',color='gold', label ='Lebron James', ax=ax)

plt.xlabel('Season Year')
plt.ylabel('Points Per Game')

plt.title ('Lebron Career Points Stats')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
```

```python
ax = plt.gca() # gca stands for 'get current axis'
bird.plot(kind='bar',x='Year',y='PPG',color='green', label ='Larry Bird', ax=ax)

plt.xlabel('Season Year')
plt.ylabel('Points Per Game')

plt.title ('Larry Legend Career Points Stats')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
```

```python
ax = plt.gca() # gca stands for 'get current axis'
kb.plot(kind='bar',x='Year',y='PPG',color='purple', label ='Kobe', ax=ax)

plt.xlabel('Season Year')
plt.ylabel('Points Per Game')

plt.title ('Kobe Bryant Career Points Stats')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
```

```python
ax = plt.gca() # gca stands for 'get current axis'
vc.plot(kind='bar',x='Year',y='PPG',color='blue', label ='Vince Carter', ax=ax)

plt.xlabel('Season Year')
plt.ylabel('Points Per Game')

plt.title ('Vince Carter Career Points Stats')

plt.rcParams['figure.figsize'] = (10,6)
plt.legend()
```

According to these Stats, Michael Jordan had the greatest Rookie year compare to all 5 players. Second best Player was Larry Bird in his Rookie year, Third best player is Lebron,
follow by Vince Carter, and lastly Kobe Bryant. All 5 players got the Rookie of the Year
Award.
